```python
import numpy as np
import pandas as pd

# Se carga Hoja de calculo
file_path = 'Calculos.xlsx'
excel_data = pd.ExcelFile(file_path)

# hoja específica "inflation_annual_percent" (solo hay 1)
sheet_data = excel_data.parse('inflation_annual_percent')

# Filtrar los datos de inflación para Colombia
colombia_data = sheet_data[sheet_data['country'] == 'Colombia'].drop(columns='country').T
colombia_data.columns = ['inflation_rate']
colombia_data.index = pd.to_numeric(colombia_data.index, errors='coerce')
colombia_data['inflation_rate'] = pd.to_numeric(colombia_data['inflation_rate'], errors='coerce')
colombia_data = colombia_data.dropna()

x = colombia_data.index.values.astype(float)
y = colombia_data['inflation_rate'].values

# Calcular el polinomio de  segundo grado
coefficients = np.polyfit(x, y, 2)
polynomial = np.poly1d(coefficients)

# Derivar el polinomio para encontrar los puntos críticos f'
derivative = np.polyder(polynomial)
critical_points = np.roots(derivative)

# Clasificar los puntos críticos en máximos y mínimos evaluando la segunda derivada
second_derivative = np.polyder(derivative)
classified_points = {point: 'Máximo' if second_derivative(point) < 0 else 'Mínimo' for point in critical_points}

print("Coeficientes del polinomio: \n", coefficients)
print("Puntos críticos y su clasificación \n", classified_points)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# codigo con ajustes para mostrar la grafica con ajustes
# Cargar el archivo Excel
file_path = 'Calculos.xlsx'
excel_data = pd.ExcelFile(file_path)

sheet_data = excel_data.parse('inflation_annual_percent')

# Filtrar los datos de inflación para Colombia
colombia_data = sheet_data[sheet_data['country'] == 'Colombia'].drop(columns='country').T
colombia_data.columns = ['inflation_rate']
colombia_data.index = pd.to_numeric(colombia_data.index, errors='coerce')
colombia_data['inflation_rate'] = pd.to_numeric(colombia_data['inflation_rate'], errors='coerce')
colombia_data = colombia_data.dropna()

x = colombia_data.index.values.astype(float)
y = colombia_data['inflation_rate'].values

# Calcular el polinomio lineal (recta)
coefficients = np.polyfit(x, y, 1)  # grado 1
slope, intercept = coefficients

# Ecuación de la recta
print(f"Ecuación de la recta: y = {slope:.4f}x + {intercept:.4f}")

# Graficar los datos y la línea de tendencia
plt.scatter(x, y, color='blue', label='Datos de Inflación')
plt.plot(x, slope * x + intercept, color='red', label='Línea de Ajuste')
plt.xlabel('Año')
plt.ylabel('Inflación (%)')
plt.title('Ajuste Lineal de la Inflación en Colombia')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Calculos.xlsx'
excel_data = pd.ExcelFile(file_path)
sheet_data = excel_data.parse('inflation_annual_percent')

def get_country_data(sheet_data, country_name):
    country_data = sheet_data[sheet_data['country'] == country_name].drop(columns='country').T
    if not country_data.empty:
        country_data.columns = ['inflation_rate']
        country_data.index = pd.to_numeric(country_data.index, errors='coerce')
        country_data['inflation_rate'] = pd.to_numeric(country_data['inflation_rate'], errors='coerce')
        return country_data.dropna()
    else:
        print(f"No se encontraron datos para {country_name}.")
        return None
colombia_data = get_country_data(sheet_data, 'Colombia')
mexico_data = get_country_data(sheet_data, 'Mexico')
china_data = get_country_data(sheet_data, 'China')

degree = 6

# Crear la gráfica
plt.figure(figsize=(15, 10))

def plot_country_data(country_data, color, label):
    if country_data is not None:
        x = country_data.index.values.astype(float)
        y = country_data['inflation_rate'].values
        coefficients = np.polyfit(x, y, degree)
        polynomial = np.poly1d(coefficients)
        plt.plot(x, y, color=color, label=f'Inflación en {label}')
        plt.plot(x, polynomial(x), color=color, linestyle='--', label=f'Ajuste Polinómico {label} (Grado {degree})')
        print(f"Ecuación del polinomio de {label}: {polynomial}")

# Graficar cada país
plot_country_data(colombia_data, 'blue', 'Colombia')
plot_country_data(mexico_data, 'green', 'Mexico')
plot_country_data(china_data, 'red', 'China')

plt.xlabel('Año')
plt.ylabel('Inflación (%)')
plt.title(f'Ajuste Polinómico de Grado {degree} para la Inflación')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Calculos.xlsx'
excel_data = pd.ExcelFile(file_path)

sheet_data = excel_data.parse('inflation_annual_percent')
def get_country_data(sheet_data, country_name):
    country_data = sheet_data[sheet_data['country'] == country_name].drop(columns='country').T
    if not country_data.empty:
        country_data.columns = ['inflation_rate']
        country_data.index = pd.to_numeric(country_data.index, errors='coerce')
        country_data['inflation_rate'] = pd.to_numeric(country_data['inflation_rate'], errors='coerce')
        return country_data.dropna()
    else:
        print(f"No se encontraron datos para {country_name}.")
        return None

# Procesar datos para cada país
colombia_data = get_country_data(sheet_data, 'Colombia')
mexico_data = get_country_data(sheet_data, 'Mexico')
china_data = get_country_data(sheet_data, 'China')

# Crear listas para los promedios
countries = []
averages = []
colors = ['blue', 'green', 'red']

# Calcular promedios y preparar datos para la gráfica de torta
if colombia_data is not None:
    countries.append('Colombia')
    averages.append(colombia_data['inflation_rate'].mean())

if mexico_data is not None:
    countries.append('Mexico')
    averages.append(mexico_data['inflation_rate'].mean())

if china_data is not None:
    countries.append('China')
    averages.append(china_data['inflation_rate'].mean())

# Crear la gráfica de torta
plt.figure(figsize=(8, 8))
plt.pie(averages, labels=countries, colors=colors[:len(countries)], autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Promedio de Inflación por País')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Calculos.xlsx'
excel_data = pd.ExcelFile(file_path)

sheet_data = excel_data.parse('inflation_annual_percent')

def get_country_data(sheet_data, country_name):
    country_data = sheet_data[sheet_data['country'] == country_name].drop(columns='country').T
    if not country_data.empty:
        country_data.columns = ['inflation_rate']
        country_data.index = pd.to_numeric(country_data.index, errors='coerce')
        country_data['inflation_rate'] = pd.to_numeric(country_data['inflation_rate'], errors='coerce')
        return country_data.dropna()
    else:
        print(f"No se encontraron datos para {country_name}.")
        return None
colombia_data = get_country_data(sheet_data, 'Colombia')
mexico_data = get_country_data(sheet_data, 'Mexico')
china_data = get_country_data(sheet_data, 'China')

#SE TIENE EL AÑO 200  se filtra el año
def filter_from_year(data, year):
    if data is not None:
        return data[data.index >= year]
    return None

colombia_2000 = filter_from_year(colombia_data, 2000)
mexico_2000 = filter_from_year(mexico_data, 2000)
china_2000 = filter_from_year(china_data, 2000)


countries_2000 = []
averages_2000 = []
colors = ['blue', 'green', 'red']
#Se promedian los años desde el año 2000
if colombia_2000 is not None:
    countries_2000.append('Colombia')
    averages_2000.append(colombia_2000['inflation_rate'].mean())

if mexico_2000 is not None:
    countries_2000.append('Mexico')
    averages_2000.append(mexico_2000['inflation_rate'].mean())

if china_2000 is not None:
    countries_2000.append('China')
    averages_2000.append(china_2000['inflation_rate'].mean())


plt.figure(figsize=(8, 8))
plt.pie(
    averages_2000,
    labels=countries_2000,
    colors=colors[:len(countries_2000)],
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)
plt.title('Promedio de Inflación por País (2000 en adelante)')
plt.axis('equal')
plt.show()
```
